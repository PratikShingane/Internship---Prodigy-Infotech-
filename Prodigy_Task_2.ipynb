##Univariate Analysis

```python
plt.figure(figsize=(8,5))
cancer['Level'].value_counts().plot (kind='bar',color=['c','darkcyan','teal'])
plt.title('Value counts of Levels column',fontsize=20,color='darkslategray')
plt.ylabel('value counts',fontsize=10,c='darkred')
plt.xlabel('Levels',fontsize=10,c='darkred')
plt.text(0,367,'365',color='red',fontweight='bold')
plt.text(1,333,'332',color='red',fontweight='bold')
plt.text(2,304,'303',color='red',fontweight='bold')
plt.grid()
plt.show()
```

Explaination :-

The risk level associated with the patient's health  condition varrys in High Medium Low
to know the specific value count of each unique value the bar graph has been used
high = 365    patient with high risk level
Medium  = 332    patient with Medium  risk level
low = 303    patient with low risk level

##KDE(Kernel density estimation)

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=cancer,x='Smoking',fill=True,color='teal',bw_adjust=1)
plt.xticks(range(0,11))
plt.grid()
plt.show()

plt.show()
```

Explaination :-
         KDE represents the data using a continuous probability density curve in one or more dimensions.
         The above KDE plot shows the probability density of 'Smoking' values
         The ratings on X-axis indicate level of smoking ranging from " 1 to 8 " with higher value indicating more smoking
         As shown in the graph
         2 , 7 , 1 & 3 level of smoking have higher density of smokers

##Bivariate Analysis

Clubbing of Finger Nails vs chronic Lung Disease

```python
ax = sns.countplot(data = cancer, x = 'Clubbing of Finger Nails', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
club_finger =cancer.groupby(['Clubbing of Finger Nails'], as_index=False)['chronic Lung Disease'].sum().sort_values(by='chronic Lung Disease', ascending=False)

sns.barplot(x = 'Clubbing of Finger Nails',y= 'chronic Lung Disease' ,data =  club_finger )

plt.show()
```

From above graphs we can see that Clubbing of Finger Nails level 2,4,5&9 male have Chronic lung Disease

###Smoking

```python
smoke_chest = cancer.groupby(['Smoking'], as_index=False)['Chest Pain'].sum().sort_values(by='Chest Pain', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data =smoke_chest , x = 'Smoking',y= 'Chest Pain')

plt.show()
```

```python
smoke_drycough = cancer.groupby(['Smoking'], as_index=False)['Dry Cough'].sum().sort_values(by='Dry Cough', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data =smoke_drycough , x = 'Smoking',y= 'Dry Cough')

plt.show()
```

From above graphs we can see that Smoking is the cause of chest pain and Dry cough

##Multivariate analysis

Coughing blood

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = cancer, x = 'Coughing of Blood')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
cough_blood = cancer.groupby(['Coughing of Blood','Level'], as_index=False)['Chest Pain'].sum().sort_values(by='Chest Pain', ascending=False)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data = cough_blood, x = 'Coughing of Blood',y= 'Chest Pain',hue='Level')

plt.show()
```

From above graphs we can see that There is higher degree of chest pain and coughing of blood in patient level 6,7,8,9

```python
sns.heatmap(cancer[['Air Pollution','Alcohol use','Dust Allergy','OccuPational Hazards','Genetic Risk','chronic Lung Disease','Balanced Diet','Obesity','Smoking','Passive Smoker','Chest Pain','Coughing of Blood']].corr(),annot=True,cmap='hot_r',cbar=True)

plt.show()
```

##Conclusion

In conclusion, the analysis highlights key connections between air pollution and lung cancer risk. It reveals a significant correlation, even among nonsmokers, emphasizing the impact of environmental factors on cancer development. Those living in areas with high pollution levels are particularly vulnerable, with nonsmokers at a heightened risk compared to smokers. Age also plays a crucial role, with risk increasing as individuals grow older. While these findings shed light on the issue, further research is essential to validate and explore these associations fully. Overall, the study underscores the importance of addressing air pollution to mitigate its harmful effects on public health, particularly concerning lung cancer incidence.
