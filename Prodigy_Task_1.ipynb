Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population

```python
# Importing library

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the CSV file

d= pd.read_csv(r'/content/Diwali Sales Data.csv',encoding='latin')
d
```

```python
plt.figure(figsize=(30,10))
sns.set(style='white')
p=sns.barplot(data =d, x ='State' ,y= 'Age',hue = 'Age Group',errorbar=None)
plt.title('Distribution of Age group as per State',size = 20)
plt.grid()
plt.show()
```
